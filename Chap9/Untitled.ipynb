```python
import matplotlib.pyplot as plt
import math
import numpy as np

def a921(t,x):
    return 1.5 * x

def b921(t,x):
    return x

def wiener_process(start,end,n):
    t = list(np.linspace(start,end,n+1))
    w =[0]
    sigma = math.sqrt(t[1]-t[0])
    for tn in t:
        w.append(w[-1]+np.random.normal(0,sigma))
    w.pop()
    return [t,w]

def euler_maruyama_sch(a,b,x0,p,delta):
    t =  list(np.linspace(p[0][0],p[0][-1],int(1/delta)+1))
    w = p[1]
    y =[x0]
    d = dict(zip(p[0],w))
    l = len(t)
    for i in range(l-1):
        y.append(y[-1]+a(t[i],y[-1])*(t[i+1]-t[i])+b(t[i],y[-1])*(d[t[i+1]]-d[t[i]]))
    return [t,y]

#def answer921():
    
w = wiener_process(0.0,1.0,2**9)
x0 = 1.0
ans = euler_maruyama_sch(a921,b921,x0,w,2**(-8))
x=[]
for a in zip(w[0],w[1]):
    t = a[0]
    wt = a[1]
    xt =  math.exp(t+wt)
    x.append(xt)
plt.plot(w[0],x)
plt.plot(ans[0],ans[1])
plt.show()
```
