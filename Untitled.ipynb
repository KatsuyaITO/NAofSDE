# ε-貪欲法
- パラメータ ε>0
- 入力:T>0
- 手順
    - すべてのアームをεT/K回引く
    - その平均が一番大きかったアームを残りの(1-ε)T回引く

```python
eps=0.1
K=1000
T=100000
num=int(eps*T/K)

print("Regret(T) OF episilon-greedy")
for n_time in range(10):
    slots={i:np.random.normal(loc=np.random.normal(),scale=np.abs(np.random.normal()),size=T) for i in range(K)}
    argmax_i=np.argmax([value[:num].mean() for (key,value) in slots.items()])
    print((np.max([value.mean() for (key,value) in slots.items()])-slots[argmax_i].mean())*T)

```

# UCB 方策
- 手順
-  すべてのアームを1回づつ引く
- for t=K+1,...,T
    + 各アームのUCBスコアを計算する
    + UCBスコアが最大のアームを引く
 

```python
# UCB方策がどのスロットを選んだかを描画する
# T=1000のK=100腕バンディット問題

K=100
T=1000

for n_time in range(10):
    slots={i:list(np.random.normal(loc=np.random.normal(),scale=np.abs(np.random.normal()),size=T)) for i in range(K)}
    Is=[]
    UCB = [[slots[i].pop()]for i in range(K)]
    for n_time1 in range(T-K):
        argmax_i = np.argmax([np.mean(ucb)+np.sqrt(np.log(T)/(2*len(ucb))) for ucb in UCB])
        Is.append(argmax_i)
        UCB[argmax_i].append(slots[argmax_i].pop())
    plt.plot(Is)
    plt.show()
    plt.close()
    
```

# ベルヌーイ分布のトンプソン抽出

- 手順
- $n_i$,$m_i$=0とする.(i=1,...,K)
- for t=1,2,...,T:
    + $\mu_i$をBeta(α+m_i,β+n_i-m_i)からランダムに生成する
    + $i=argmax(\mu_i)$
    + アームiを引いて,X_iを観測する
    + n_i<-n_i+1,m_i<-m_i+X_i



```python
K=100
T=1000

n=[0 for i in range(K)]
m=[0 for i in range(K)]
for n_time in range(10):
    slots={i:list(np.random.beta(
                a=np.abs(np.random.normal()),
                b=np.abs(np.random.normal()),
                size=T)) for i in range(K)}
    Is=[]
    for t in range(T):
        mu_ = [np.random.beta(1+m[i],1+n[i]-m[i])for i in range(K)]
        i=np.argmax(mu_)
        Is.append(i)
        X_t = slots[i].pop()
        n[i]+=1
        m[i]+=X_t
    plt.plot(Is)
    plt.show()
    plt.close()
    
```
